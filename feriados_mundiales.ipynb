```python
import pandas as pd
import numpy as np
import io
import boto3
import holidays
import warnings
warnings.filterwarnings('ignore')
import pycountry
from pycountry import countries
```

```python
df_country = pd.DataFrame()
x=0
all_countries = list(pycountry.countries)
for country in all_countries:
    df_country.at[x,'country_id'] = country.alpha_2
    df_country.at[x,'country'] = country.name
    x+=1
```

```python
def get_holidays(country, year):
    return holidays.CountryHoliday(country, years=year)
```

```python
#countries = ['AR','US','PE','UY','MX','CO','VE','JP','CH']
year = 2023
```

```python
df_holidays = pd.DataFrame()

for country in df_country.country_id.to_list():
    try:
    
        df_pre = pd.DataFrame.from_dict(get_holidays(country, year).items()).reset_index(drop=True)

        df_pre.columns = ['date','name']

        df_pre['country_id'] = country

        df_holidays = df_holidays.append(df_pre).reset_index(drop=True)

        print(f"Holidays in {country} in {year} Done")
        
    except:
        
        print(f"Country {country} not available")
```

```python
df_holidays = df_holidays.merge(df_country,on='country_id',how='left')
```
